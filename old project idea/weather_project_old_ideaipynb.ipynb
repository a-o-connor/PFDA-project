```python
mullingar_daily_weather= pd.read_csv("dly875.csv", skiprows=23)
mullingar_daily_weather = mullingar_daily_weather[["date","rain", "maxtp", "mintp", "wdsp"]]
mullingar_daily_weather["date"] = pd.to_datetime(mullingar_daily_weather["date"], format='%d-%b-%Y')
mullingar_daily_weather.set_index("date", inplace=True)
mullingar_daily_weather
```

```python
dublinairport_daily_weather= pd.read_csv("dly532.csv", skiprows=24)
dublinairport_daily_weather = dublinairport_daily_weather[["date","rain", "maxtp", "mintp", "wdsp"]]
dublinairport_daily_weather["date"] = pd.to_datetime(dublinairport_daily_weather["date"], format='%d-%b-%Y')
dublinairport_daily_weather.set_index("date", inplace=True)
dublinairport_daily_weather
```

```python
#Plot the temperature data over time: 
fig, ax = plt.subplots(2,2,figsize=(20, 12))

sns.lineplot(
    mullingar_daily_weather["rain"],
    ax=ax[0,0]
)
ax[0,0].set_ylabel("Daily Rainfall (mm)")
ax[0,0].set_xlabel("Date")
ax[0,0].set_title("Rainfall Data for Mullingar 1974-2024")

sns.lineplot(
    mullingar_daily_weather["wdsp"],
    ax=ax[0,1]
)
ax[0,1].set_ylabel("Daily Windspeed (kt)")
ax[0,1].set_xlabel("Date")
ax[0,1].set_title("Windspeed Data for Mullingar 1974-2024")

sns.lineplot(
    mullingar_daily_weather["maxtp"],
    ax=ax[1,0]
)
ax[1,0].set_ylabel("Max. Daily Temperature (°C)")
ax[1,0].set_xlabel("Date")
ax[1,0].set_title("Temperature Data for Mullingar 1974-2024")
```
